# **Project Summary -**

The primary objective of this project was to explore, clean, analyze, and visualize financial data in order to uncover patterns, customer behavior, and risk indicators that can guide effective business decision-making. The dataset contained key attributes such as **Age, Occupation, Annual Income, Credit Score, Credit Mix, and Outstanding Debt**, all of which directly influence creditworthiness and repayment capacity.

The project began with **data wrangling**, which involved identifying missing values, handling duplicates, correcting data types, and ensuring consistency across variables. Unique values for categorical attributes such as `Occupation` and `Credit_Mix` were inspected to understand data distribution, while continuous variables like `Age`, `Annual_Income`, and `Outstanding_Debt` were checked for outliers and anomalies. This step ensured data reliability and laid a strong foundation for subsequent analysis.

### Exploratory Data Analysis (EDA) and Visualizations

A variety of visualization techniques were employed to extract meaningful insights:

1. **Histograms and Boxplots** were used to analyze **Age** and **Annual Income distributions**. These charts highlighted that most customers fall within the working-age group, but a significant proportion has low to moderate incomes. The presence of income outliers was also detected, suggesting uneven earning potential within the dataset.

2. **Scatterplots of Age vs Outstanding Debt** provided a deeper look at how different age groups manage debt. Younger individuals (<25 years) often carried disproportionately higher outstanding balances compared to their income levels, signaling a risk of default due to unstable financial stability.

3. **Bar plots for Credit Score and Credit Mix** revealed variations in financial discipline. A sizable portion of customers possessed poor credit scores, and many relied heavily on unsecured credit options such as credit cards. These factors indicate risky borrowing behavior that could adversely affect loan repayment rates.

4. **Occupation vs Credit Score Countplots** highlighted that certain professions, such as students and freelancers, were more likely to have weaker credit profiles. On the other hand, salaried professionals displayed stronger repayment capacity and higher credit scores.

5. **Correlation Heatmaps** were utilized to understand relationships between numerical variables. Interestingly, **Annual Income and Outstanding Debt** showed weak correlation, indicating that higher income does not necessarily translate to lower debt. This suggests overspending tendencies among customers, regardless of income levels.

### Key Insights

From these analyses, several positive and negative business implications were drawn:

* **Positive Impacts:**

  * Middle-aged, salaried professionals form a stable customer base with healthier credit scores and manageable debt levels.
  * Income segmentation allows financial institutions to design targeted loan products, improving customer satisfaction and profitability.
  * Clear identification of risk-prone groups enables businesses to strengthen credit policies and reduce defaults.

* **Negative Growth Indicators:**

  * Younger customers with high debt and low repayment capacity increase the risk of non-performing assets (NPAs).
  * High reliance on unsecured credit signals weak financial discipline, threatening loan recovery rates.
  * Income inequality and weak correlation between income and debt suggest unsustainable borrowing behavior.

### Business Impact

The insights from this project provide actionable strategies for financial institutions. By focusing on **low-risk customer segments**, designing differentiated loan products, and tightening approval criteria for high-risk groups, businesses can achieve **sustainable growth**. At the same time, addressing negative growth factors through stricter risk assessment, customer financial education, and personalized repayment options can help mitigate potential losses.

### Conclusion

Overall, this project successfully demonstrated the role of **data wrangling, exploratory analysis, and visualization** in deriving insights that directly influence financial decision-making. The findings highlight both opportunities and risks within the customer base, providing a balanced perspective that can help businesses optimize lending strategies, reduce credit risk, and enhance profitability. With proper action on these insights, financial institutions can strengthen their portfolio quality, minimize defaults, and ensure long-term business growth.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# 1. Droping irrelevant columns
drop_cols = ["ID", "Customer_ID", "Name", "SSN", "Month"]
data.drop(columns=drop_cols, inplace=True, errors='ignore')
```

```python
# Filling numeric columns with median
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    data[col] = data[col].fillna(data[col].median())

# Filling categorical columns with mode
for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
# Standardizing categorical values
if "Payment_of_Min_Amount" in data.columns:
    data["Payment_of_Min_Amount"] = data["Payment_of_Min_Amount"].replace(
        {"NM": "No", "NO": "No", "N": "No", "YES": "Yes", "Y": "Yes"}
    )
```

```python
if "Credit_Mix" in data.columns:
    data["Credit_Mix"] = data["Credit_Mix"].replace(
        {"_": "Unknown", "nan": "Unknown"}
    )
```

```python
# Encode Credit Score categories
score_mapping = {"Poor": 0, "Standard": 1, "Good": 2}
data['Credit_Score_Encoded'] = data['Credit_Score'].map(score_mapping)
```

```python
# 5. Final check
print("\nFinal dataset shape:", data.shape)
print("Columns after cleaning:\n", data.columns.tolist())
```

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Income affects Credit Score

Null Hypothesis (H₀): There is no significant difference in Annual Income across different Credit Score categories.

Alternative Hypothesis (H₁): There is a significant difference in Annual Income across different Credit Score categories.

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import f_oneway

# Group incomes by credit score
good_income = data[data['Credit_Score']=="Good"]["Annual_Income"]
std_income = data[data['Credit_Score']=="Standard"]["Annual_Income"]
poor_income = data[data['Credit_Score']=="Poor"]["Annual_Income"]

# ANOVA test
f_stat, p_val = f_oneway(good_income, std_income, poor_income)
print("ANOVA p-value:", p_val)
```

##### Which statistical test have you done to obtain P-Value?

I used One-Way ANOVA (Analysis of Variance) to compare the mean Annual Income across the three Credit Score groups (Good, Standard, Poor).

##### Why did you choose the specific statistical test?

ANOVA is used when you want to compare a numeric variable across more than two categorical groups.

In this case, Annual_Income is numeric and Credit_Score has 3 categories → so ANOVA is the correct test.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Payment Behaviour is independent of Credit Score

H₀: Payment Behaviour and Credit Score are independent.

H₁: Payment Behaviour and Credit Score are associated (i.e., they influence each other).

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import chi2_contingency

# Create contingency table
contingency = pd.crosstab(data['Payment_Behaviour'], data['Credit_Score'])

chi2, p, dof, expected = chi2_contingency(contingency)
print("Chi-square p-value:", p)
```

##### Which statistical test have you done to obtain P-Value?

I used the Chi-Square Test of Independence to examine whether Payment Behaviour and Credit Score are related.

##### Why did you choose the specific statistical test?

Both Payment_Behaviour and Credit_Score are categorical variables.

The Chi-Square test checks if the distribution of one variable is independent of the other.

This makes it the correct choice to test association between payment discipline and credit scores.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Outstanding Debt is correlated with Credit Utilization Ratio

H₀: There is no significant correlation between Outstanding Debt and Credit Utilization Ratio.

H₁: There is a significant correlation between Outstanding Debt and Credit Utilization Ratio.

#### 2. Perform an appropriate statistical test.

```python
from scipy.stats import pearsonr

corr, p_val = pearsonr(data['Outstanding_Debt'], data['Credit_Utilization_Ratio'])
print("Correlation coefficient:", corr)
print("p-value:", p_val)
```

##### Which statistical test have you done to obtain P-Value?

I used the Pearson Correlation Test to measure the linear correlation between Outstanding Debt and Credit Utilization Ratio.

##### Why did you choose the specific statistical test?

Both Outstanding_Debt and Credit_Utilization_Ratio are continuous numeric variables.

Pearson correlation quantifies the strength and direction of a linear relationship between two continuous variables.

Therefore, it is the correct test to determine if higher debt is associated with higher utilization.

### 2. Handling Outliers

```python
for col in ['Annual_Income', 'Outstanding_Debt']:
    lower = data[col].quantile(0.01)
    upper = data[col].quantile(0.99)
    data[col] = np.clip(data[col], lower, upper)
```

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used Quantile Capping (Winsorization at 1st and 99th percentiles) to handle extreme outliers in Annual Income and Outstanding Debt. This ensures the model is not dominated by unrealistic values while still retaining all customer records.

### 3. Categorical Encoding

```python
data.select_dtypes(include=['object']).columns
```

```python
# Binary encoding
data['Payment_of_Min_Amount'] = data['Payment_of_Min_Amount'].map({'Yes': 1, 'No': 0})

# Ordinal encoding
mix_mapping = {"Bad": 0, "Standard": 1, "Good": 2}
data['Credit_Mix_Encoded'] = data['Credit_Mix'].map(mix_mapping)

score_mapping = {"Poor": 0, "Standard": 1, "Good": 2}
data['Credit_Score_Encoded'] = data['Credit_Score'].map(score_mapping)
```

I used Ordinal Encoding for Credit_Mix and Credit_Score because they have a natural order (Bad < Standard < Good, Poor < Standard < Good). This preserves ranking information so the model can correctly interpret creditworthiness levels.

I used Binary Encoding for Payment_of_Min_Amount (Yes → 1, No → 0) because it is a binary categorical feature. This simple numeric mapping makes the data efficient to process without adding extra columns.

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['Annual_Income', 'Outstanding_Debt', 'Monthly_Balance', 'Monthly_Inhand_Salary',
            'Credit_Utilization_Ratio', 'Debt_to_Income_Ratio', 'Savings_Ratio']

scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
```

##### Which method have you used to scale your data and why?

StandardScaler was chosen over MinMaxScaler because the dataset contained outliers. StandardScaler is more robust in such cases, while MinMaxScaler could have been heavily skewed by extreme values

### 8. Data Splitting

```python
# One-hot encode Occupation & Payment_Behaviour
data = pd.get_dummies(data, columns=['Occupation', 'Payment_Behaviour'], drop_first=True)

# Drop the target from features
X = data.drop(columns=['Credit_Score_Encoded'])
y = data['Credit_Score_Encoded']

# Split again with stratification
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Check object columns (should be empty now)
print(X_train.select_dtypes(include=['object']).columns)
```

##### What data splitting ratio have you used and why?

I used an 80:20 train-test split with stratification on the target variable.

Why 80:20?

80% for training → provides enough data for the model to learn underlying patterns.

20% for testing → ensures a fair and independent evaluation of model performance.

This ratio is widely used in industry as a good balance between learning and validation.
